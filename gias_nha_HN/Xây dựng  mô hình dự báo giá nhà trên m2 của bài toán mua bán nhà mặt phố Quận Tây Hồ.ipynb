```python
# điền dữ liệu khuyết thiếu

# cột diện tích
df1['dien_tich'].fillna(df1['gia'] / df1['gia_m2'], inplace = True);

# cột giá
df1['gia'].fillna(df1['dien_tich'] * df1['gia_m2'], inplace = True);

# cột giá trên/m2
df1['gia_m2'].fillna(df1['gia'] / df1['dien_tich'], inplace = True);

# cột số phòng ngủ 

mapping = {1:1, 2:2, 3:3, 4:5, 5:6, 6:7, 7:9, 8:13, 9:14, 10:17, 11:28, 23:255}
df1['phong_ngu'] = df1['phong_ngu'].fillna(df1['so_tang'].map(mapping));
# df1.loc[(df1['so_tang'] >= 1) & (df1['so_tang'] <= 3), 'phong_ngu'].fillna(3, inplace=True)
# df1.loc[(df1['so_tang'] == 4), 'phong_ngu'].fillna(5, inplace = True)
# df1.loc[(df1['so_tang'] == 5), 'phong_ngu'].fillna(6, inplace = True)
# df1.loc[(df1['so_tang'] == 6), 'phong_ngu'].fillna(7, inplace = True)
# df1.loc[(df1['so_tang'] == 7), 'phong_ngu'].fillna(9, inplace = True)
# df1.loc[(df1['so_tang'] == 8), 'phong_ngu'].fillna(13, inplace = True)
# df1.loc[(df1['so_tang'] == 9), 'phong_ngu'].fillna(14, inplace = True)
# df1.loc[(df1['so_tang'] == 10), 'phong_ngu'].fillna(17, inplace = True)
# df1.loc[(df1['so_tang'] == 11), 'phong_ngu'].fillna(28, inplace = True)
# df1.loc[(df1['so_tang'] == 23), 'phong_ngu'].fillna(225, inplace = True)

# cột số tầng
def fill_so_tang(dataframe):
    for index, row in dataframe.iterrows():
        if row['phong_ngu'] == 1:
            if np.isnan(row['so_tang']):
                dataframe.loc[index, 'so_tang'] = 1
                
        elif row['phong_ngu'] > 2 and row['phong_ngu'] <= 3:
            if np.isnan(row['so_tang']):
                dataframe.loc[index, 'so_tang'] = 2
                
        elif row['phong_ngu'] > 3 and row['phong_ngu'] <= 5:
            if np.isnan(row['so_tang']):
                dataframe.loc[index, 'so_tang'] = 3
                
        elif row['phong_ngu'] > 5 and row['phong_ngu'] <= 8:
            if np.isnan(row['so_tang']):
                dataframe.loc[index, 'so_tang'] = 4
                
        elif row['phong_ngu'] > 8 and row['phong_ngu'] <= 11:
            if np.isnan(row['so_tang']):
                dataframe.loc[index, 'so_tang'] = 5
                
        elif row['phong_ngu'] > 11 and row['phong_ngu'] <= 14:
            if np.isnan(row['so_tang']):
                dataframe.loc[index, 'so_tang'] = 6
        
        elif row['phong_ngu'] > 14 and row['phong_ngu'] <= 17:
            if np.isnan(row['so_tang']):
                dataframe.loc[index, 'so_tang'] = 7
        
        elif row['phong_ngu'] > 17 and row['phong_ngu'] <= 20:
            if np.isnan(row['so_tang']):
                dataframe.loc[index, 'so_tang'] = 8
        
        elif row['phong_ngu'] > 20 and row['phong_ngu'] <= 25:
            if np.isnan(row['so_tang']):
                dataframe.loc[index, 'so_tang'] = 9
        
        elif row['phong_ngu'] > 25 and row['phong_ngu'] <= 30:
            if np.isnan(row['so_tang']):
                dataframe.loc[index, 'so_tang'] = 10
    return dataframe
fill_so_tang(df1);

# cột mặt tiền
df1['mat_tien'].interpolate(method='linear',
                            inplace=True,
                            limit_direction='both',
                            limit_area='inside',
                            downcast='infer',
                            x=df1['dien_tich'])
# cột độ rộng đường
df1['do_rong_duong'].interpolate(method='linear',
                            inplace=True,
                            limit_direction='both',
                            limit_area='inside',
                            downcast='infer');

# cột độ rộng mặt đường
def convert_street_width(_str):
    index_lst = ['Mặt phố - Mặt đường',
                 'Ngõ 4 ô tô trở lên',
                 'Ngõ 3 ô tô tránh',
                 'Ngõ 1 ô tô',
                 'Ngõ 4 ô tô tránh',
                 'Ngõ 2 ô tô tránh',
                 'Ngõ ngách']
    if pd.isna(_str):
        return 0
    elif(_str in index_lst):
        return index_lst.index(_str)
df1['do_rong_duong_ml'] = df1['do_rong_duong_ml'].apply(lambda x: convert_street_width(x));
    
# cột sổ đỏ    
def convert_policy(_str):
    if pd.isna(_str):
        return 0
    return 1
df1['so_do'] = df1['so_do'].apply(lambda x: convert_policy(x))
```

```python
df3 = df1.dropna()
x = (df3.shape[0]/df1.shape[0]) *100
y = round(x,2)
print(' số % giá trị còn lại so với ban đầu:', y,'%')
```

```python
# xử lý giá trị ngoại lai

def get_q1_q3(values):
    values = np.array(values)
    q1 = np.quantile(values, 0.25)
    q3 = np.quantile(values, 0.75)
    iqr = q3 - q1
    _min_value = q1 - 1.5*iqr
    _max_value = q3 + 1.5*iqr
    
    return _min_value, _max_value

min_dien_tich, max_dien_tich = get_q1_q3(df3['dien_tich'])

min_so_tang, max_so_tang = get_q1_q3(df3['so_tang'])

min_phong_ngu, max_phong_ngu = get_q1_q3(df3['phong_ngu'])

min_mat_tien, max_mat_tien = get_q1_q3(df3['mat_tien'])

min_do_rong_duong, max_do_rong_duong = get_q1_q3(df3['do_rong_duong'])

min_gia, max_gia = get_q1_q3(df3['gia'])

min_gia_m2, max_gia_m2 = get_q1_q3(df3['gia_m2'])


logic = ((df3['dien_tich'] >= min_dien_tich) & (df3['dien_tich'] <= max_dien_tich) & 
         (df3['so_tang'] >= min_so_tang) & (df3['so_tang']<= max_so_tang) & 
         (df3['phong_ngu']>=min_phong_ngu) & (df3['phong_ngu']<= max_phong_ngu) & 
         (df3['do_rong_duong']>=min_do_rong_duong ) & (df3['do_rong_duong']<= max_do_rong_duong ) & 
         (df3['mat_tien']>=min_mat_tien) & (df3['mat_tien']<= max_mat_tien) & 
         (df3["gia"]>=min_gia) & (df3["gia"]<= max_gia) & 
         (df3["gia_m2"] >= min_gia_m2) & (df3["gia_m2"]<= max_gia_m2))

clean = df3[logic]
clean
```

```python
# split data
X = clean[['dien_tich', 'so_tang', 'phong_ngu', 'mat_tien','do_rong_duong', 'do_rong_duong_ml','so_do', 'gia']]
y = clean["gia_m2"]
X_train, X_test, y_train, y_test = train_test_split(
   X, y, test_size=0.25, random_state=42)

model = linear_model.LinearRegression()
model.fit(X_train, y_train)

# đánh giá metrics
y_pred = model.predict(X_test)
mse = mean_absolute_error(y_pred, y_test)
mape = mean_absolute_percentage_error(y_pred, y_test)
print(mse)
print(mape)
```

```python
# Lưu mô hình vào file
with open('linear_regression_model.pkl', 'wb') as file:
    pickle.dump(model, file)
```
