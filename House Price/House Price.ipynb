```python
# check null

def find_missing_percent(data):
    miss_df = pd.DataFrame({'col':[], 'total':[], 'percent':[]})
    for col in data:
        miss_info = {'col': col,
                     'total': data[col].isna().sum()}
        miss_df = miss_df.append(miss_info, ignore_index = True)
    miss_df = miss_df[miss_df['total'] > 0]
    miss_df['percent'] = round(miss_df['total'] / data.shape[0]*100, 2)
    return miss_df
```

```python
# approach 2: missing handling with mean/median/mode imputation

def mean_imputation(data, miss_num_col):
    for col in miss_num_col:
        data[col] = data[col].fillna(data[col].mean())
    return data
def mode_imputation(data, miss_cat_col):
    for col in miss_cat_col:
        data[col] = data[col].fillna(data[col].mode().iloc[0])
    return data
c2 = mean_imputation(data, miss_num_col)
c2 = mode_imputation(data, miss_cat_col)
```

```python
# approach 03: missing handling with imputation by grouping

import association_metrics as am
df = data[cat_col].copy()
df = df.apply(lambda x: x.astype('category') if x.dtype == 'O' else x)
cramersv = am.CramersV(df)
cate_corr_matrix = cramersv.fit()
```

```python
def group_cat_imputation(data, cate_corr_matrix, miss_cat_col):
    group_dict = {}
    for col in miss_cat_col:
        corr_col = cate_corr_matrix[col].sort_values(ascending = False).keys()[1]
        rowIdx_notnull = data[corr_col].notnull()
        rowIdx_null = data[col].isnull()
        targetdict = {val : data[data[corr_col] == val][col].mode().iloc[0]
                      for val in data.loc[rowIdx_notnull, corr_col].unique()}
        group_dict[col] = {corr_col: targetdict}
        data['Map_Value'] = data[corr_col].map(targetdict)
        data.loc[rowIdx_null, col] = data.loc[rowIdx_null, 'Map_Value']
        
        rowIdx_null_all = data[col].isnull() & data[corr_col].isnull()
        data.loc[rowIdx_null_all, col] = data.loc[rowIdx_null_all, col].fillna(data[col].mode().iloc[0])
    return data, group_dict
```

```python
data_cate_imp_group, group_dict = group_cat_imputation(data, cate_corr_matrix, miss_cat_col)
```

```python
def NormalizeSkewedFeatures(data_modelling):
    from scipy.special import boxcox1p
    from scipy.stats import boxcox_normmax
    for col in skew_cols.index:
        if(col != 'SalePrice'):
            try:
                data_modelling[col] = boxcox1p(data_modelling[col], boxcox_normmax(data_modelling[col] + 1))
            except:
                print(f"column {col} can not apply BoxCox")
                continue
    return data_modelling
```

```python
def DummyEncoding(data_modelling, cat_col):
    data_modelling_dummy = pd.get_dummies(data_modelling, columns=cat_col, drop_first=True)   
    return data_modelling_dummy

def FrequencyEncoding(data_modelling, cat_col):
    data_modelling_frequency = data_modelling.copy()
    for col in cat_col:
        fq = data_modelling_frequency.groupby(col).size()/len(data_modelling_frequency)   
        data_modelling_frequency.loc[:, col] = data_modelling_frequency[col].map(fq) 
    return data_modelling_frequency
```

```python
from sklearn.linear_model import Ridge, RidgeCV, Lasso, LassoCV
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error, r2_score

def DataSplitTrainTest(data_modelling):
    train = data_modelling.copy()
    X = train.drop('SalePrice', axis=1)
    y = train['SalePrice']
    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=12345)
    print("Train Data", X_train.shape)
    print("Test Data", X_test.shape)
    return X_train, X_test, y_train, y_test
```

```python
def BuildLassoModel(X_train, X_test, y_train, y_test):
    lasso = Lasso(max_iter = 100000)
    lassocv = LassoCV(alphas = None, cv = 10, max_iter = 100000)
    lassocv.fit(X_train, y_train)

    lasso.set_params(alpha=lassocv.alpha_)
    lasso.fit(X_train, y_train)

    print('The Lasso Regression Result:')
    print("\t Alpha =", lassocv.alpha_)
    print("\t RMSE =", mean_squared_error(y_test, lasso.predict(X_test), squared=False))
    print("\t R2 Score = ", r2_score(y_test, lasso.predict(X_test)))
    return lasso
```

```python
def BuildRidgeModel(X_train, X_test, y_train, y_test):
    alphas = np.geomspace(1e-9, 5, num=100)
    ridgecv = RidgeCV(alphas = alphas, scoring = 'neg_mean_squared_error', normalize = True)
    ridgecv.fit(X_train, y_train)

    ridge = Ridge(alpha = ridgecv.alpha_, normalize = True)
    ridge.fit(X_train, y_train)

    print('The Ridge Regression Result:')
    print("\t Alpha =", ridgecv.alpha_)
    print("\t RMSE =", mean_squared_error(y_test, ridge.predict(X_test), squared=False))
    print("\t R2 Score = ", r2_score(y_test, lasso.predict(X_test)))
    return ridge
```
